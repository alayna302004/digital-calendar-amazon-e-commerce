# Digital Calendar Amazon E-Commerce Data Analysis - by Alayna Smith

```python
import pandas as pd
import numpy as np
import altair as alt
```

## Original Dataset:

```python
df = pd.read_csv('Digital Calendar Helium 10 Export.csv')
df
```

```python
df.info()
```

## Cleaned Dataset:

```python
excluded_columns = ['Product Details', 'URL', 'Image URL', 'Weight',
                    'Parent Level Sales', 'Parent Level Revenue',
                    'Review velocity', 'Size Tier', 'Dimensions']
filtered_df = df.drop(columns=excluded_columns)
columns_to_convert = ['ASIN Sales', 'ASIN Revenue', 'BSR', 'Review Count',
                      'Price  $', 'Title Char. Count', 'Recent Purchases']

for column in columns_to_convert:
  # Replace 'N/A' with np.nan (NaN represents null in pandas)
  filtered_df[column] = filtered_df[column].replace('N/A', np.nan)
  # Remove commas from the string values
  filtered_df[column] = filtered_df[column].astype(str).str.replace(',', '', regex=False)
  # Convert to float, coercing errors to NaN
  filtered_df[column] = pd.to_numeric(filtered_df[column], errors='coerce')
# Fill empty 'Category' with 'Unknown'
filtered_df['Category'].fillna('Unknown', inplace=True)
# Fill empty 'Sponsored' with 'None'
filtered_df['Sponsored'].fillna('None', inplace=True)

filtered_df.info()
```

```python
filtered_df
```

## First Visualization: Revenue vs. Brand (top 10 brands)

```python
# top ten brands, based on ASIN revenue
top_rev_brands = filtered_df.groupby('Brand')['ASIN Revenue'].sum().nlargest(10).reset_index()
top_rev_brands
```

```python
chart1 = alt.Chart(top_rev_brands).mark_bar().encode(
    x=alt.X('Brand', sort='-y'),
    y='ASIN Revenue',
    color='Brand'
).properties(
    width=600,
    height=500,
    title='Top 10 Brands by ASIN Revenue'
)
chart1.show()
```

## Second Visualization: Revenue vs. Category

```python
# Group by 'Category' and sum 'ASIN Revenue'
category_revenue = filtered_df.groupby('Category')['ASIN Revenue'].sum().reset_index()
category_revenue
```

```python
chart2 = alt.Chart(category_revenue).mark_bar().encode(
    x=alt.X('Category', sort='-y'),
    y='ASIN Revenue',
    color='Category'
  ).properties(
    width=600,
    height=500,
    title='ASIN Revenue by Category'
  )
chart2
```

## Third Visualization: Revenue vs. Product Price
